```python
def Amdahl(N, P):
    print(1 - P, P / N, N)
    
    print(1/(1 - P), 1/(P / N), N)
    return 1/((1 - P) + P / N) 
    
def Gustafson(N, P):
    S = []
    for n, p in zip(N, P):
        S.append(1 + (n-1)*p)
    return S
    
```
