```python
import matplotlib.pyplot as plt
import numpy as np

from utils import *
```

```python
partial_store_threads, partial_store_mem = get_pStore_data()
print(partial_store_threads)

average_total_times, average_worst_times, average_total_std, average_worst_2_std  = calculate_avg_and_worst_times("pStore_threads", "time_threads", partial_store_threads)
# server_median, server_min, server_max, server_std = load_cpu_stats_from_files_p√Ænned("other/pinned_server", system_matrix_size)

print(average_total_times)
print(average_worst_times)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Plot CPU with error bars
ax.errorbar(
    partial_store_threads, average_total_times, yerr=average_total_std,
    marker='o', label='average thread', color='blue', capsize=5, linestyle='-'
)

# Plot GPU with error bars
ax.errorbar(
    partial_store_threads, average_worst_times, yerr=average_worst_2_std,
    marker='s', label='worst thread', color='orange', capsize=5, linestyle='-'
)

ax.set_xlabel('amount of threads', fontsize=24)
ax.set_ylabel('Execution \nTime (s)', fontsize=24)
# ax.set_title('Execution Time vs threads', fontsize=24)

ax.tick_params(axis='both', labelsize=24)
# ax.set_yscale('log')

fig.canvas.draw()
offset_text = ax.xaxis.get_offset_text()
offset_text.set_size(24)

ax.legend(fontsize=24)

plt.ylim(bottom=0)
ax.grid(True)
plt.tight_layout()
plt.savefig("graphs/pStore_threads.svg", format='svg')
plt.show()
```
