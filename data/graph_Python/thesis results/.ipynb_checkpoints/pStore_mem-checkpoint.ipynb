```python
import matplotlib.pyplot as plt
import numpy as np

from utils import *
```

```python
partial_store_threads, partial_store_mem = get_pStore_data()
print(partial_store_mem)

average_total_times, average_worst_times, average_total_std, average_worst_2_std = calculate_avg_and_worst_times("pStore_mem", "time_mem", partial_store_mem)
# server_median, server_min, server_max, server_std = load_cpu_stats_from_files_p√Ænned("other/pinned_server", system_matrix_size)

memory_also_used = (5898240000 + 30105600) / (2**30)


partial_store_mem = partial_store_mem - memory_also_used# in GiB
print(partial_store_mem)

print(average_total_times)
print(average_worst_times)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Plot CPU with error bars
ax.errorbar(
    partial_store_mem, average_total_times, yerr=average_total_std,
    marker='o', label='average thread', color='blue', capsize=5, linestyle='-'
)

# Plot GPU with error bars
ax.errorbar(
    partial_store_mem, average_worst_times, yerr=average_worst_2_std,
    marker='s', label='worst thread', color='orange', capsize=5, linestyle='-'
)

ax.set_xlabel('size of pinned memory (GiB)', fontsize=24)
ax.set_ylabel('Execution \nTime (s)', fontsize=24)
# ax.set_title('Execution Time vs threads', fontsize=24)

ax.tick_params(axis='both', labelsize=24)
# ax.set_yscale('log')

fig.canvas.draw()
offset_text = ax.xaxis.get_offset_text()
offset_text.set_size(24)

ax.legend(fontsize=24)

plt.ylim(bottom=0)
ax.grid(True)
plt.tight_layout()
plt.savefig("graphs/pStore_mem.svg", format='svg')
plt.show()
```

```python
partial_store_threads, partial_store_mem = get_pStore_data()
average_total_times, average_worst_times, average_total_std, average_worst_std = calculate_avg_and_worst_times_components("pStore_mem", "time_mem", partial_store_mem)

memory_also_used = (5898240000 + 30105600) / (2**30)


partial_store_mem = partial_store_mem - memory_also_used# in GiB
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Plot CPU with error bars
ax.errorbar(
    partial_store_mem, average_total_times[:, 0], yerr=average_total_std[:, 0],
    marker='o', label='average thread', color='blue', capsize=5, linestyle='-'
)

# Plot GPU with error bars
ax.errorbar(
    partial_store_mem, average_worst_times[:, 0], yerr=average_worst_std[:, 0],
    marker='s', label='worst thread', color='orange', capsize=5, linestyle='-'
)

ax.set_xlabel('amount of threads', fontsize=24)
ax.set_ylabel('Execution \nTime (s)', fontsize=24)
# ax.set_title('Execution Time vs threads', fontsize=24)

ax.tick_params(axis='both', labelsize=24)
# ax.set_yscale('log')

fig.canvas.draw()
offset_text = ax.xaxis.get_offset_text()
offset_text.set_size(24)

ax.legend(fontsize=24)

plt.ylim(bottom=0)
ax.grid(True)
plt.tight_layout()
plt.savefig("graphs/pStore_mem_calculate.svg", format='svg')
plt.show()
```
