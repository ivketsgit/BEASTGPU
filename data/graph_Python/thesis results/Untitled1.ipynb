```python
A = np.array(range(1, 6))
n = 76800**2 / A
```

```python
plt.figure(figsize=(12, 6))
plt.plot(A, n, 'o--')
plt.xlabel('Number of workitems',fontsize=24)
plt.ylabel('Speedup',fontsize=24)
plt.title("Amdahl's Law",fontsize=24)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.legend()
plt.axis([0, None, 0, None])
plt.grid(True)
plt.tight_layout()
#plt.xscale('log')
plt.show()
```
