```python
speedup_CPU_main  = CPU_main[4] / ( integral[:, 0] + overhead[:, 0])
speedup_CPU_V     = CPU_V[4] / ( integral[:, 1] + overhead[:, 1])
speedup_CPU_E     = CPU_E[4] / (integral[:, 2]  + overhead[:, 2])
speedup_CPU_F     = CPU_F[4] /  (integral[:, 3]  + overhead[:, 3])
# print(speedup_CPU_E)
```

```python
speedup_matrix = np.column_stack([
    speedup_CPU_main,
    speedup_CPU_V,
    speedup_CPU_E,
    speedup_CPU_F
])

cases = [
    # (CPU_main[4], integral[:, 0] + overhead[:, 0]),
    (CPU_V[4],    integral[:, 1] + overhead[:, 1]),
    (CPU_E[4],    integral[:, 2] + overhead[:, 2]),
    (CPU_F[4],    integral[:, 3] + overhead[:, 3])
]

cols = []
for cpu_val, gpu_time in cases:
    speedup = cpu_val / gpu_time
    cols.append(np.full_like(gpu_time, cpu_val))
    cols.append(gpu_time)                         
    cols.append(speedup)                        

full_matrix = np.column_stack(cols)


latex_matrix = "\\begin{bmatrix}\n"
for row in full_matrix:
    latex_matrix += " & ".join(f"{val:.3f}" for val in row) + r" \\" + "\n"
latex_matrix += "\\end{bmatrix}"

print(latex_matrix)
```
